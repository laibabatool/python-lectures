### 1. Defining a function
A function is a block of organized, reusable code that is used to perform a single, related action. Functions provide better modularity for your application and a high degree of code reusing.

To define a function, you use the `def` keyword followed by the function name, parentheses `()`, and a colon `:`. The function body is indented.

```python
def greet(): # define
    """This function prints a simple greeting."""
    print("Hello, World!")

# Calling the function
greet()
```

### 2. Arguments vs. Parameters
- **Parameters**: The names listed inside the parentheses in the function definition.
- **Arguments**: The actual values passed to the function when it is called.

```python
def introduce(name, age): # name and age are parameters
    print(f"My name is {name} and I am {age} years old.")

introduce("Alice", 30) # "Alice" and 30 are arguments
```

### 3. All types of arguments
Python supports several types of arguments:

#### a) Positional Arguments
The order of arguments matters. The first argument passed corresponds to the first parameter, and so on.

```python
def subtract(a, b):
    return a - b

print(subtract(10, 5))  # a=10, b=5 -> 5
print(subtract(5, 10))  # a=5, b=10 -> -5
```

```python
def fullname(fname,lname):
    print(fname + lname)

fullname("laiba","batool")
fullname("batool","laiba")
```

#### b) Keyword Arguments
Arguments are passed using the parameter names. The order does not matter.

```python
def divide(numerator, denominator):
    return numerator / denominator

print(divide(numerator=10, denominator=2)) # order doesn't matter
print(divide(denominator=2, numerator=10))
```

```python
def fullname(fname,lname):
    print(fname + lname)

fullname(fname = "laiba",lname="batool")
fullname(lname="batool",fname="laiba")
```

#### c) Default Arguments
Parameters can have default values. If an argument is not provided for such a parameter, its default value is used.

```python
# optional/default arguments
def greet_person(name="Guest", message="Hello"):
    print(f"{message}, {name}!")

greet_person() # Uses default values
greet_person("Bob") # Overrides name
greet_person("Charlie", "Hi") # Overrides both
greet_person(message="Good morning") # Overrides message with keyword
```

```python
def fullname(fname,lname = "Khan"): #default argument "lname"
    print(fname + lname)

fullname("laiba ")
fullname("laiba ","batool")
fullname("Hira ")
```

### 4. `*args` and `**kwargs`
These allow functions to accept an arbitrary number of arguments.

#### a) `*args` (Arbitrary Positional Arguments)
Collects all extra positional arguments into a tuple.

```python
# *args Arbitrary positional arguments
def sum_all(*numbers,*):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_all(1, 2, 3))       # numbers = (1, 2, 3)
print(sum_all(10, 20, 30, 40)) # numbers = (10, 20, 30, 40)
print(sum_all(1,1,1))
```

```python
def sum_all(number1,number2):
    return number1 + number2

sum_all(2,5)
```

#### b) `**kwargs` (Arbitrary Keyword Arguments)
Collects all extra keyword arguments into a dictionary.

```python
def show_details(**details):
    for key, value in details.items():
        print(f"{key.replace('_', ' ').title()}: {value}")

show_details(name="Alice", age=30, city="New York")
print("\n---")
show_details(product="Laptop", price=1200, brand="Acme Corp")
```

### 5. Scope of Variables
Variable scope refers to where a variable is accessible within a program.

#### a) Local Scope
Variables defined inside a function are local to that function and cannot be accessed from outside it.

```python
def my_function():
    local_var = "I am local"
    print(local_var)

my_function()
# print(local_var) # This would raise a NameError
```

#### b) Global Scope
Variables defined outside any function are global and can be accessed (and modified with the `global` keyword) from anywhere in the program.

```python
global_var = "I am global"

def another_function():
    print(global_var) # Can access global_var

def modify_global():
    global global_var # Declare intent to modify the global variable
    global_var = "I was modified globally"

print(f"Initial global_var: {global_var}")
another_function()
modify_global()
print(f"Modified global_var: {global_var}")
```
