# Python Sets 

## What is a Set in Python?

A set is a built-in Python data structure that stores:
- Unique values (no duplicates)
- Unordered elements
- Mutable (can change after creation)

```python
my_set = {1, 2, 3, 3, 4}
print(my_set)
```

## Creating a Set

```python
fruits = {"apple", "banana", "mango"}
print(fruits)

num_list = [1, 2, 3, 4]
numbers = set(num_list) # converting list into set
print(numbers)
```

## Empty Set

```python
empty_set = set()
print(type(empty_set))
# for creating empty set we have to use set(), we cant use {}
empty = {}
print(type(empty))
```

## Adding Elements

```python
colors = {"red", "blue","yellow","black"}
colors.add("green")
print(colors)
```

## Removing Elements

```python
colors.remove("pink") # remove element (error if missing)
print(colors)
```

```python
colors.discard("pink")  # safely remove
print(colors)
```

```python
colors.discard("red")  # safely remove
print(colors)
```

```python
# pop() removes an arbitrary element (not truly random, but unpredictable)
# You don’t control which item is removed
item = colors.pop()
print(item)
```

```python
print(colors)
```

```python
colors.clear() # deletes all the elements
print(colors)
```

## Set Operations

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B) # Union : Give me all unique values from both groups
print(A & B) # intersection: Give me what both sets have in common
print(A - B) # difference: Remove all elements of B from A
print(A ^ B) # Symmetric Difference: Give me everything except the common part
```

## Membership Test

```python
numbers = {10, 20, 30}
print(20 in numbers)
print(50 in numbers)
```

## Loop Through a Set

```python
for item in {1, 2, 3}:
    print(item)
```

## List vs Tuple vs Set

- List:
  - Ordered
  - Mutable
  - Allows duplicates
  - Supports indexing

- Tuple:
  - Ordered
  - Immutable
  - Allows duplicates
  - Faster than list

- Set:
  - Unordered
  - Mutable
  - No duplicates allowed
  - No indexing
  - Very fast for membership checks
