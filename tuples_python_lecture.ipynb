# Tuples in Python

##  Definition
A **tuple** is an ordered, immutable (unchangeable) collection of elements in Python.


##  List vs Tuple
- List: Mutable, uses []
- Tuple: Immutable, uses ()


```python
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
print(my_list, my_tuple)
```

##  Creating Tuples

```python
t1 = (1, 2, 3)
t2 = 1, 2, 3
t3 = (5,)
t4 = (1, "hello", 3.5, True)
print(t1, t2, t3, t4)
```

```python
# in Python, it’s the comma , that creates a tuple, not the parentheses.

t = (5)    # NOT a tuple
print(type(t))  # int

t = (5,)   # tuple
print(type(t))  # tuple
```

##  Tuple Methods

```python
t = (1, 2, 2, 3, 2)
print("Count of 2:", t.count(2))
print("Index of 3:", t.index(3))
```

##  Accessing Elements

```python
t = (10, 20, 30, 40)
print(t[0])
print(t[-1])
print(t[1:3])
```

##  Updating Tuple (Workaround)

```python
t = (1, 2, 3)
temp = list(t)
temp[0] = 10
t = tuple(temp)
print(t)
```

```python
temp1 = list(t)
temp1.append(5)
t = tuple(temp1)
print(t)
```

##  Iterating Tuples

```python
t = (1, 2, 3)
for i in t:
    print(i)

```

##  Packing & Unpacking

```python
# Packing means putting multiple values into a single variable (usually a tuple).

t = 1, 2, 3  # packing
a, b, c = t  # unpacking
print(a, b, c)
```

```python
a, *b = (1, 2, 3, 4)  # advanced unpacking: *b means, remaining items into b as a list
print(a, b)
```

##  Exercises

```python
# Exercise 1 : Create a tuple and print its first and last element.
t = (10, 20, 30, 40)
print(t[0], t[-1])
print("-------------------------------------------")

# Exercise 2 : Count how many times 5 appears.
t = (1, 5, 3, 5, 5)
print(t.count(5))
print("-------------------------------------------")

# Exercise 3 : Convert tuple into list and update value.
t = (1, 2, 3)
l = list(t)
l[1] = 10
t = tuple(l)
print(t)
print("-------------------------------------------")

# Exercise 4 : Iterate through tuple and print squares.
t = (1, 2, 3, 4)
for i in t:
    print(i*i)
print("-------------------------------------------")

# Exercise 5 : Unpack tuple into variables.
t = (100, 200, 300)
a, b, c = t
print(a, b, c)
print("-------------------------------------------")

# Exercise 6 :Find maximum value in tuple.
t = (4, 7, 2, 9, 1)
print(max(t))
```

```python
# Why Tuple is Faster
# Tuple is immutable (no changes allowed) → less processing
# Uses less memory → faster access
# No need for resizing or updating → more efficient
# Has fewer built-in methods → lightweight


# Why List is Slower
# List is mutable (can change) → extra processing needed
# Uses more memory → slightly slower
# Needs to handle insert, delete, resize operations
# Has many built-in methods → more overhead
```
